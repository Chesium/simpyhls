```python
from compiler.workflow import (
    compile_file,
    write_compilation_outputs,
)
from compiler.primitive_rtl import PrimitiveRTLRegistry, PrimitiveRTLSpec

from pathlib import Path

ROOT = Path(".").resolve()
luCorePath = ROOT / "examples" / "lu_core.dsl.py"
luCoreOutputPath = ROOT / "tmp" / "lu_core.sv"
luCoreHIRPath = ROOT / "tmp" / "lu_core.hir.txt"
luCoreLIRPath = ROOT / "tmp" / "lu_core.lir.txt"
luCoreReportPath = ROOT / "tmp" / "lu_core.report.txt"


def lu_registry() -> PrimitiveRTLRegistry:
    return PrimitiveRTLRegistry(
        [
            PrimitiveRTLSpec(name="neg_comb", ports=("v")),
            PrimitiveRTLSpec(
                name="fetch_A", ports=("i", "j"), result_port="result", latency=1
            ),
            PrimitiveRTLSpec(
                name="fetch_LU", ports=("i", "j"), result_port="result", latency=1
            ),
            PrimitiveRTLSpec(name="store_LU", ports=("i", "j", "v"), latency=1),
            PrimitiveRTLSpec(
                name="fma", ports=("a", "b", "c"), result_port="result", latency=16
            ),
            PrimitiveRTLSpec(
                name="div", ports=("a", "b"), result_port="result", latency=28
            ),
        ]
    )


compilation_artifact = compile_file(luCorePath, lu_registry())
write_compilation_outputs(
    compilation_artifact,
    luCoreOutputPath,
    hir_path=luCoreHIRPath,
    lir_path=luCoreLIRPath,
    report_path=luCoreReportPath,
)
```

```python
floodingCorePath = ROOT / "examples" / "flooding_core.dsl.py"
floodingCoreOutputPath = ROOT / "tmp" / "flooding_core.sv"
floodingCoreHIRPath = ROOT / "tmp" / "flooding_core.hir.txt"
floodingCoreLIRPath = ROOT / "tmp" / "flooding_core.lir.txt"
floodingCoreReportPath = ROOT / "tmp" / "flooding_core.report.txt"


def flooding_registry() -> PrimitiveRTLRegistry:
    return PrimitiveRTLRegistry(
        [
            PrimitiveRTLSpec(
                name="fetchP", ports=("i", "j"), result_port="result", latency=1
            ),
            PrimitiveRTLSpec(name="storeR", ports=("i", "j", "v"), latency=1),
            PrimitiveRTLSpec(
                name="getVisited", ports=("i", "j"), result_port="result", latency=1
            ),
            PrimitiveRTLSpec(name="setVisited", ports=("i", "j"), latency=1),
            PrimitiveRTLSpec(name="addQueue", ports=("i", "j", "d"), latency=1),
            PrimitiveRTLSpec(
                name="popQueue", ports=(), result_port="result", latency=1
            ),
            PrimitiveRTLSpec(
                name="getQueueLen", ports=(), result_port="result", latency=1
            ),
            PrimitiveRTLSpec(name="decode_iswire_comb", ports=("p")),
            PrimitiveRTLSpec(name="getport_comb", ports=("p", "i")),
            PrimitiveRTLSpec(name="decode_i_comb", ports=("q_item")),
            PrimitiveRTLSpec(name="decode_j_comb", ports=("q_item")),
            PrimitiveRTLSpec(name="decode_d_comb", ports=("q_item")),
            PrimitiveRTLSpec(name="get_nxt_i_comb", ports=("i", "d")),
            PrimitiveRTLSpec(name="get_nxt_j_comb", ports=("j", "d")),
            PrimitiveRTLSpec(name="get_opp_dir_comb", ports=("d")),
        ]
    )


compilation_artifact = compile_file(floodingCorePath, flooding_registry())
write_compilation_outputs(
    compilation_artifact,
    floodingCoreOutputPath,
    hir_path=floodingCoreHIRPath,
    lir_path=floodingCoreLIRPath,
    report_path=floodingCoreReportPath,
)
```
